# [Thực hành] Truy cập dữ liệu trong DataFrame

image.png

1. Đọc dữ liệu

```python
import pandas as pd
df=pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31021019/FoodPrice_in_Turkey.csv')
df.head()
```

2. Sử dụng Iloc để truy cập dữ liệu

```python
# Truy cập 1 dòng
iloc_1 = df.iloc[4] # Lấy ở dòng thứ 5 với dòng có chỉ số = 4
iloc_1
```

```python
# Truy cập các dòng có chỉ số liên tục từ 3 tới 8 của dữ liệu
iloc_2 = df.iloc[3:8]
iloc_2
```

```python
# Truy cập các dòng rời rạc của dữ liệu
iloc_3 = df.iloc[[3,5,7]]
iloc_3
```

```python
# Truy cập cột thứ 4 của dữ liệu
# : được sử dụng để đại diện cho tất cả các dòng.
iloc = df.iloc[:,4]
iloc
```

```python
# Truy cập các cột liên tục từ cột 3 tới cột 8 của dữ liệu(bao gồm cột 3,4,5,6,7)
tg = df.iloc[:,3:8]
tg
```

```python
# Truy cập cột thứ 3 của dữ liệu
# truy cập 5 dòng đầu của cột thứ 3
iloc = df.iloc[:5,3]
iloc
```

```python
# Truy cập cột thứ 6 của dữ liệu
# truy cập dòng 3,4,8 đầu của cột thứ 6
iloc = df.iloc[[3,4,8],6]
iloc
```

```python
# Truy cập tới các phần tử từ dòng 3 đến dòng 4 của cột 5 đến cột 6 của dữ liệu
tg = df.iloc[3:5,5:7]
tg
```

3. Sử dụng phương thức loc để truy cập dữ liệuloc có thể làm việc được với chỉ số hàng, tên cột và các biểu thức điều kiện của cột

```python
# Truy cập dòng có chỉ số 3 của dữ liệu
tg = df.loc[3]
tg
```

```python
# Truy cập cột thứ 4 của dữ liệu
tg = df.loc[:,'UmName']
tg
```

```python
# Truy cập cột thứ 4,5 của dữ liệu
tg = df.loc[:,['UmName','Month']]
tg
```

```python
# Truy cập 6 dòng của cột thứ 4,5 của dữ liệu
tg = df.loc[:5,['UmName','Month']]
tg
```

```python
# Truy cập dòng 1,3,5 của cột thứ 4,5 của dữ liệu
tg = df.loc[[1,3,5],['UmName','Month']]
tg
```

```python
# Truy cập tới phần tử tại dòng 3 cột 7 của dữ liệu
tg = df.loc[3,'Price']
tg
```

```python
# Truy cập tới phần tử có Year >=2019
tg = df.loc[df.Year >= 2019]
tg
```

4. Thay thế các giá trị trong DataFrame

```python
df.replace(5,10, inplace=True) # Thay số 5 bằng số 10 trong toàn bộ dữ liệu
df.head()
```

```python
df.replace(2013,2022, inplace=True) # Thay năm 2013 =  năm 2022 trong toàn bộ dữ liệu
df.head()
```

```python
# Thay mã sản phẩm từ giá trị 52 thành DA trong toàn bộ dữ liệu
# DA là chuỗi ký tự nên được để trong cặp dấu nháy đơn
df.replace(52,'DA',inplace = True)
df.head()
```

```python
# Thay giá trị 10 trong cột Month thành giá trị 5
df['Month'].replace(10,5,inplace = True)
df.head()
```

```python
# Thay giá trị KG trong cột UmName thành giá trị SA
df['UmName'].replace('KG','SA',inplace = True)
df.head()
```
