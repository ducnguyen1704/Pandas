Import thư viện

```python
import pandas as pd
import numpy as np
```

1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
df = pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31011319/OnlineRetail.csv', encoding= 'unicode_escape')
df.head(10)
```

```python
df.info()
```

- Bộ dữ liệu gồm 541909 dòng và 8 cột
- Có 2 cột mang kiểu dữ liệu số thực là UnitPrice và CustomerID
- CÓ 1 cột có kiểu dữ liệu nguyên là Quantity
- Có 5 cột có kiểu dữ liệu chuỗi là InvoiceNo , StockCode, Description, InvoiceDate và Country

2. Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

```python
df_2 = df.loc[:,['Description','Quantity']]
df_2
```

```python
# Lưu vào file OnlineRetail.csv
df_2.to_csv('OnlineRetail.csv')
```

3. Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx

```python
df_3 = df.iloc[:1000]
df_3
```

```python
# Lưu vào file excel
df_3.to_excel('OnlineRetail.xlsx')
```

4. Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5

```python
df_4 = df.loc[df['Quantity'] >= 10]
df_4
```

```python
# Lưu file h5
df_4.to_hdf('OnlineRetail.h5','table')
```

image.png

5. Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json

```python
df_5 = df.loc[1000:2000,['Quantity','InvoiceDate','UnitPrice']]
df_5
```

```python
# Lưu vào file json
df_5.to_json('OnlineRetail.json')
```

6. Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html

```python
df_6 = df[df.InvoiceNo == '536365']
df_6
```

```python
# Lưu file html
df_6.to_html('OnlineRetail.html')
```
