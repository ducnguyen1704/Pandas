# [Thực hành] Ghi dữ liệu từ DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31021019/FoodPrice_in_Turkey.csv')
data.head()
```

1. Ghi dữ liệu từ DataFrame vào file csv

```python
# Ghi dữ liệu từ DataFrame vào file csv
data.to_csv('demo_FoodPrice.csv')
```

Kết quả được lưu lại <br>
image.png

```python
# đọc file vừa lưu
demo = pd.read_csv('demo_FoodPrice.csv')
demo.head()
```

2. Ghi dữ liệu từ DataFrame vào file excel

```python
data.to_excel('demo_FoodPrice.xlsx')
```

```python
# đọc file vừa lưu
demo = pd.read_excel('demo_FoodPrice.xlsx')
demo.head()
```

3. Ghi dữ liệu từ DataFrame vào file Json <br>
Tham số orient được sử dụng để xác định định dạng dữ liệu đầu ra, có thể là:

+ Các bản ghi – records
+ Các cột – columns
+ Các chỉ mục – indexs
+ Các giá trị – values

```python
data.to_json('demo_FoodPrice.json',orient='columns')
```

```python
# đọc file vừa lưu
demo = pd.read_json('demo_FoodPrice.json')
demo.head()
```

4. Ghi dữ liệu từ DataFrame vào file HDF5

```python
data.to_hdf('demo_FoodPrice.h5', 'table')
```

5. Ghi dữ liệu từ DataFrame vào file HTML

```python
data.to_html('demo_FoodPrice.html')
```
